```r
#see if there are differences in body mass between male and female penguins
p = ggplot(penguins, aes(x=body_mass_g, y=sex, color=species)) + geom_point(position=position_jitter(height=0.2, width=0)) +scale_color_brewer(palette = 'Set 2')

p
#executing ggplot and capturing it as variable p, then calling p
```

```r
options(repr.plot.height=5, repr.plot.width=4)

p = ggplot(penguins |> drop_na(), aes(x=body_mass_g, y = sex, color=species)) +
geom_point(alpha=0.5, size = 1, position=position_jitter(height=0.3, width=0)) +
facet_wrap(~species, ncol=1) + scale_color_brewer(palette = 'Set 2') + theme_bw()

p

# + signs have to be on first line, then enter. line cannot begin with +
#overwrite the value of p here with new chart
#jupyter works in order you execute the cells, not top down. can restart and run top to bottom to reset everything 
```
